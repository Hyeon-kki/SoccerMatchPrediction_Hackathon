```python
train, valid = load_train_valid()

homeTeam = list(train['homeTeam'].unique())
awayTeam = list(train['awayTeam'].unique())
homeWindic = {}
awayWindic = {}
totalWindic = {}
for team in homeTeam:
    homeWindic[team] = 0
    awayWindic[team] = 0
    totalWindic[team] = 0

home_idx = train[train['result'] == 'H'].index
away_idx = train[train['result'] == 'A'].index

for idx in home_idx:
    team = train.loc[idx]['homeTeam']
    homeWindic[team] = homeWindic[team]+1

for idx in away_idx:
    team = train.loc[idx]['awayTeam']
    awayWindic[team] = awayWindic[team]+1

for team, value in homeWindic.items():
    totalWindic[team] = homeWindic[team] + awayWindic[team]

sorted_team = sorted(totalWindic.items(), key=lambda item: item[1])

range_400_500 = []
range_180_400 = []
range_100_180 =[]
range_50_100 =[]
others =[]
for series in sorted_team:
    n = series[1]
    if n > 400:
        range_400_500.append(series[0])
    elif n>180 and n<400 :
        range_180_400.append(series[0])
    elif n>100 and n<180:
        range_100_180.append(series[0])
    elif n>50 and n<100:
        range_50_100.append(series[0])
    else:
        others.append(series[0])

def team_category(team):
    if team in range_400_500:
        return 0
    elif team in range_180_400:
        return 1
    elif team in range_100_180:
        return 2
    elif team in range_50_100:
        return 3
    elif team in others:
        return 4
```
