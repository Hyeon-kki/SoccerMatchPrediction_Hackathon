```python
import pandas as pd
import numpy as np
import random
import os
import lightgbm as lgb

import statsmodels.api as sm
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import log_loss

import sys
sys.path.append("..")
from utils import *
from preprocessing_utils import *
import warnings
warnings.filterwarnings("ignore")

def seed_everything(seed):
    random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    np.random.seed(seed)

seed_everything(42)
```

```python
# 너무  잘게 쪼개니깐 성능이 부서짐

train, valid = load_train_valid()
valid.reset_index(drop= True, inplace= True)

hometeam_list = list(train['homeTeam'].unique())
dic = team_encoding(train)

lec = LabelEncoder()
lec.fit(train['result'])

result_array = np.zeros((len(valid), 3))
multi_logloss_sum = 0

for team in hometeam_list:
    team_train = train[train['homeTeam'] == team]
    team_valid = valid[valid['homeTeam'] == team]
    valid_idx = team_valid.index.values
    train_x, train_y, valid_x, valid_y  = preprocessing(team_train, team_valid, dic, is_test=False)
    train_y = lec.transform(train_y)
    valid_y = lec.transform(valid_y)

    if len(valid_x) != 0:
        model = LogisticRegression(max_iter=100, penalty='l2', C=1.0)
        model.fit(train_x, train_y) 
        prediction = model.predict_proba(valid_x)
        result_array[valid_idx] = prediction

    else:
        continue
```

```python
multi_loloss = log_loss(valid['result'], result_array)
multi_loloss
```
